ASSIGNMENT AFTER LESSON FOR FLOWCHART FOR CLASS AVERAGE

```python
#Initialization
grades=[98,76,71,87,83,90,57,79,82,94]
total=0
counter=0
#processing stage

for grade in grades:
  total+=grade
  counter +=1
average=total/counter
print(f'Class average is {average}')
```
